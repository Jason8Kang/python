### and vs &

**and**	用法：x and y,	布尔"与" - 如果 x 为 False，x and y 返回 False，否则它返回 y 的计算值。如(a and b) 返回 20。

**&**	用法：x & y，按位与运算符：参与运算的两个值,如果两个相应位都为1,则该位的结果为1,否则为0	(a & b) 输出结果 12 ，二进制解释： 0000 1100


```python
original=[str(i)+ '+'+str(j) for i in [True,False] for j in [True,False]]
result=[i and j for i in [True,False] for j in [True,False]]
dict(zip(original,result))
```

```python
original=[str(i)+ '+'+str(j) for i in [True,False] for j in [True,False]]
result=[i & j for i in [True,False] for j in [True,False]]
dict(zip(original,result))
```

### pivot table

```python
# 对照excel的power_pviot，行标签取Creditability，列标签为空，数值为age
df.pivot_table(index=u'Creditability',values='Age',aggfunc=np.mean)
```

```python
#对照excel的power_pviot，行标签取Creditability，列标签为Purpose，数值为CreditAmount
df.pivot_table(index=u'Creditability',columns=u'Purpose',values=u'CreditAmount',aggfunc=np.sum)
```
